```python
import numpy as np
import pandas as pd
# 生成一个 50x5 的随机整数数组，取值范围在 50 到 100 之间（包含 50，不包含 101）
students = np.random.randint(50,101,size=(50,5))
# 创建一个 DataFrame，命名列为 '國文', '英文', '數學', '地理', '歷史'
# 命名行索引为 '第1號' 到 '第50號'
stu = pd.DataFrame(students,
             columns=['國文','英文','數學','地理','歷史'],
             index=[f'第{i}號' for i in range(1,51)]
             )
stu
```

```python
# 调用 abc 函数计算平均值
def abc(s)->float:
    return s.mean()

stu.apply(abc,axis=1)
```

```python
#定义了一个函数 efg，它接受一个参数 s，假设 s 是一个 Pandas Series 对象，然后返回一个新的 Pandas Series 对象，其中包含了输入 Series s 的均值、最大值和最小值。
# 这个新的 Series 的索引分别是 "平均"、"最高分" 和 "最低分"。

#然后，你使用了 stu.apply(efg, axis=1)，这将应用函数 efg 到 DataFrame stu 的每一行（因为 axis=1 表示按行应用），并返回一个包含计算结果的新 DataFrame。
def efg(s) -> pd.Series: #return series
    return pd.Series([s.mean(),s.max(),s.min()],
                     index=["平均","最高分","最低分"])
stu.apply(efg,axis=1)
```

```python
def efg(s) -> pd.Series: #return series
    max_name = s[s == s.max()].index.values.tolist()[0]    
    min_name = s[s == s.min()].index.values.tolist()[0]
    less60 = ",".join(list(s[s < 60].index))
    return pd.Series([s.mean(),f'{max_name}({s.max()})',f'{min_name}({s.min()})',less60],
                     index=["平均","最高分","最低分","不及格科目"])
expand_stu = stu.apply(efg,axis=1)
pd.concat([stu,expand_stu],axis=1)
```
