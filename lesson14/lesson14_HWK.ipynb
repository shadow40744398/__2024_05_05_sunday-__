```python
import matplotlib.pyplot as plt

def main(): 
        labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
        values = [20, 30, 45, 10]
        colors =['yellow', 'green', 'red', 'blue']

        explode = (0.1, 0, 0, 0)  # 仅分离第一个部分（Nokia）
        
        # 绘制饼图，使用不同参数
        figure = plt.figure(figsize=(5,5))
        axes = figure.add_subplot(1,1,1)
        axes.pie(values, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True, explode=explode)
        plt.show()

if __name__ == "__main__":
	main()
```
