```python
import requests
from requests import Response
from pprint import pprint
import pyinputplus as pyip
from random import choices
#pydantic                 #
from pydantic import BaseModel,RootModel,Field,ValidationError,field_validator
#class method,python
class UBike(BaseModel):
	sna:str # 
	sarea:str # 
	mday:str # 
	ar:str  # 
	act:bool # 
	total:int # 
	available_rent_bikes:int# 
	latitude:float  # 
	longitude:float # 
	available_return_bikes:int # 
	

	#1.RootModel  Record  RootModel 2. JSON  RootModel 
    #3.34.35.6.
	# pydantic  RootModel UBike 
class UBikeRoot(RootModel):
	root:list[UBike]

def main():
	youbike_url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'
	res:Response = requests.get(youbike_url)
	all_data = res.json()

	try:
		root:UBikeRoot = UBikeRoot.model_validate(all_data)
	except ValidationError as error:
		print(error)

	data_list:list[dict]= root.model() # 
	filter_list: list[dict] = list(filter(lambda item:False if item['act']==0 else True, data_list)) # 
	filter_list_rent_less3:list[dict] = list(filter(lambda item:True if item['']<3 else False,data_list))
	filter_list_return_less3:list[dict] = list(filter(lambda item:True if item['']<3 else False,data_list))

	print(f"YouBike{len(data_list)}{len(filter_list)}")

	num = pyip.inputInt("(1~5): ", min=1, max=5)
	print(num)

	print(f"\n3{len(filter_list_rent_less3)}")
	num_list:list[dict] = choices(filter_list_rent_less3, k=num)
	for sta in num_list:
            seq = filter_list_rent_less3.index(sta)
            print(f"No.{seq}")
            pprint(sta, indent=2, sort_dicts=False)

	print(f"\n3{len(filter_list_return_less3)}")
	num_list:list[dict] = choices(filter_list_return_less3, k=num)
	for sta in num_list:
            seq = filter_list_return_less3.index(sta)
            print(f"No.{seq}")
            pprint(sta, indent=2, sort_dicts=False)

if __name__ == '__main__':
    main()
```
