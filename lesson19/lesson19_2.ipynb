```python
import pandas as pd
import json
# 定义 merge 函数
def merge(filename:str,station_name:pd.DataFrame) -> pd.DataFrame:
    df1 = pd.read_csv(filename) # 读取CSV文件
    # 重命名CSV文件中的列
    df2 = df1.rename(columns={'trnOpDate':'日期',
                    'staCode':'車站代碼',
                    'gateInComingCnt':'進站人數',
                    'gateOutGoingCnt':'出站人數'})
    # 转换日期列为日期格式
    df2['日期'] = pd.to_datetime(df2['日期'].astype(str))
    # 合并两个数据框
    result_df = pd.merge(df2,station_name,left_on='車站代碼',right_on='編碼',how='left')
    return result_df
# 读取JSON文件
with open('每日各站進出站人數/車站基本資料集.json',encoding='utf-8') as file:
    json_data:list[dict] = json.load(file)
# 创建车站基本资料的数据框
stations_name = pd.DataFrame(json_data,columns=['stationCode','stationName','stationAddrTw'])
stations_name1 = stations_name.rename(columns={'stationCode':'編碼',
                      'stationName':'車站名稱',
                      'stationAddrTw':'地址'})
# 转换编碼列为整数类型
stations_name1['編碼'] = stations_name1['編碼'].astype(int)
# 合并数据
merge(filename='每日各站進出站人數/每日各站進出站人數2020.csv',station_name=stations_name1)
```

```python
import os
current_path = os.path.abspath(__name__) #目前執行檔案的絕對路徑
main_path = os.path.dirname(current_path) #取得目前執行的目錄的絕對路徑

'每日各站進出站人數'
dataFolder_path = os.path.join(main_path,'每日各站進出站人數') #取得放資料的目錄
filenames = os.listdir(dataFolder_path) #取得資料目錄的檔案名稱
names = list(filter(lambda name: '每日各站進出站人數' in name,filenames)) #lambda# 过滤文件名，获取包含 '每日各站進出站人數' 的文件

# 打印每个符合条件的文件的完整路径
for name in names:
    print(os.path.join(dataFolder_path,name))

abs_names = [os.path.join(dataFolder_path,name) for name in names] #list comprehension# 获取每个文件的完整路径列表

all_data = [merge(filename=abs_n,station_name=stations_name1) for abs_n in abs_names] #list comprehension# 合并所有文件的数据

result = pd.concat(all_data,ignore_index=True)# 合并所有数据到一个 DataFrame 中

result1 = result.set_index('日期')# 设置 '日期' 列为索引


result2 = result1.sort_index().reindex(columns=['車站名稱','地址','進站人數','出站人數'])# 重新排序并选择特定的列
result2
    
```
